```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
from pyspark.sql import SparkSession
from pyspark.sql.functions import min, col


spark = SparkSession.builder.getOrCreate()


df = spark.read.csv("Indian Airlines_prices.csv", header=True)

# Create a function to find the cheapest flight
def find_cheapest_flight():
    departure_city = source_city.get()
    arrival_city = destination_city.get()
    df_filtered = df.filter((col("source_city") == departure_city) & (col("destination_city") == arrival_city))

  # Group the flights by airline
    df_filtered = df_filtered.groupBy("airline","flight")

  # Find the minimum price for each group of flights
    df_filtered = df_filtered.agg(min("price"))


  # Sort the resulting DataFrame by price in ascending order
   

# Get the first row of the DataFrame, which will be the cheapest flight
    cheapest_flight = df_filtered.first()
    


  # Display the cheapest flight in the label
    result_label.config(text="The cheapest flight from %s to %s is: %s" % (departure_city, arrival_city, str(cheapest_flight)))

    
    
    
# Create the main window
root = tk.Tk()
root.title("Airolysis")
root.geometry('3000x2000')
root.configure(bg='#FFFFFF')

label = tk.Label(root, text='AIROLYSIS', font=('Times 20 italic bold', 39), fg = 'black')
label.place(relx=0.49, rely=0.3, anchor='center')

# photo = tk.PhotoImage(file="button.png")

# Create a button with the image
# button = Button(root, image=button.png)
# button.pack()


# Create the departure location combo box
source_city = ttk.Combobox(root, values=df.select("source_city").distinct().collect(), font=('Arial', 16), width=20)
source_city.place(relx=0.3, rely=0.5, anchor='center')

# Create the arrival location combo box
destination_city= ttk.Combobox(root, values=df.select("destination_city").distinct().collect(), font=('Arial', 16), width=20)
destination_city.place(relx=0.7, rely=0.5, anchor='center')


# Create the find button
find_button = tk.Button(root, text="SEARCH", command= find_cheapest_flight,bg='red', fg='white', font=('Arial', 16))
find_button.place(relx=0.5, rely=0.6, anchor='center')

# Create the result label
result_label = tk.Label(root, font = ('Arial', 12), bg = 'black', fg= 'white' )
result_label.place(relx=0.5, rely=0.7, anchor= 'center')
find_cheapest_flight()
# Run the main loop
root.mainloop()
```
